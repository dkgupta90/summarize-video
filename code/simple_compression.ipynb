```python
# importing the necessary packages
import numpy as np
import cv2
```

```python
input_video_file = "../data/videos/002/002.mp4"
output_video_file = "../data/videos/002/002_out.avi"
```

<video controls src="../data/videos/002/002.mp4">

```python
def update_residual(old_frame, frame, change_residual):
    
    rmse = np.sqrt(np.sum(np.square(old_frame - frame)) / (old_frame.shape[0] * old_frame.shape[1]))
    change_residual = change_residual + rmse
    print(change_residual)
    return change_residual
```

```python
# Reading the video frame-by-frame
cap = cv2.VideoCapture(input_video_file)

if (cap.isOpened() == False):
    print("Error opening video stream or file")

success, frame = cap.read()
frame_width = int(frame.shape[1])
frame_height = int(frame.shape[0])
out = cv2.VideoWriter(output_video_file, cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))
out.write(frame) #Writing the first frame

# Defining the change-controlling parameters
count = 0 # counting the index of the frame read
change_residual = 0
change_threshold = 10
# Reading the next frames
while success:
    old_frame = frame
    success, frame = cap.read()
    if success == False:
        break
    change_residual = update_residual(old_frame, frame, change_residual)
    
    if change_residual > change_threshold:
        print('Flagged')
        out.write(frame)
        change_residual = 0
print('Writing completed')
cap.release()
out.release()
```
